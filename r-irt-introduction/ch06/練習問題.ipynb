```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def iif3PL(a, b, c, theta):
    D = 1.7
    p = c + (1-c)*(1/(1+np.exp(-D*a*(theta-b))))
    q = 1-p
    p_prime = D*a*q*(p-c)/(1-c)
    iif = (p_prime ** 2) / (p*q)
    return iif
```

```python

A = [1.27, 1.34, 1.14, 1.00, 0.67,1.2]
B = [1.19, 0.59, 0.15, -0.59, -2.00,-1.0]
C = [0.10, 0.15, 0.15, 0.20, 0.01]
for a,b,c in zip(A,B,C):
    x = []
    y = []
    for i in range(-30,30):
        x.append(i/10)
        iif = iif3PL(a,b,c,i/10)
        y.append(iif)
    fig, ax = plt.subplots()
    ax.plot(x, y)

    ax.set(xlabel='theta', ylabel='iif',
           title='3PLM iif')
    ax.grid()
    plt.show()
```
